## Seawater pH VS Carbon Dioxide 

Information about this subject can be found through the U.S. Environmental Protection Agency's website: https://www.epa.gov/climate-indicators/climate-change-indicators-ocean-acidity

```python
# Read in data that will be used for answering questions.
data = pd.read_csv("./Ocean_Acidity_fig-1.csv")

# Observe the first 10 rows of the data provided below.
data.head(10)
```

```python
# The graphs for Hawaii are shown in green, for Carnary Islands are shown in red, and for Bermuda are shown in blue.
fig = plt.figure(figsize=(10,3))
plt.plot(data['Hawaii Year'],data['Hawaii pH'], 'g')
fig = plt.figure(figsize=(10,5))
plt.scatter(data['Hawaii pH'], data ['Hawaii pCO2'],color='g')

fig = plt.figure(figsize=(10,3))
plt.plot(data['Canary Islands Year'],data['Canary Islands pH'], 'r')
fig = plt.figure(figsize=(10,5))
plt.scatter(data['Canary Islands pH'], data['Canary Islands pCO2'],color='r')

fig = plt.figure(figsize=(10,3))
plt.plot(data['Bermuda Year'],data['Bermuda pH'], 'b')
fig = plt.figure(figsize=(10,5))
plt.scatter(data['Bermuda pH'], data['Bermuda pCO2'],color='b')
```
