# Gravitational Orbits of Exoplanets

Background information: Exoplanets are planets that orbit a star other than the Sun outside of our solar system.  Transiting exoplanets are those that block out some light from their parent star.  Transiting is similar to an eclipse, but only a small amount of light is blocked instead of the entire planet.  Scientists use the shadows on the parent star to identify exoplanets and collect data about them.  

This notebook uses data of transiting exoplanets to analyze factors affecting their gravitational orbits.  



version:  FINAL DRAFT

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# IMPORTANT CONSTANTS USED IN THIS NOTEBOOK
Mass_sun = 1.989e30                               # Kilograms
au       = 149597870700                           # 1 Astronomical Unit =  149597870700 (m)
day      = 86400                                  # Seconds
G        = 6.67e-11*(au**-3)*(Mass_sun)*(day**2)  # Gravitational_Constant  
                                                  # Original units(m^3 kg^-1 s^-2)
                                                  # Converted to (au^3 solar_mass^-1 earth_days^-2)

```

## Transiting Exoplanets Data
Using the .csv file downloaded from http://exoplanets.org/csv-files/exoplanets.1436203548.csv.gz, a data table was generated below and used to graph the plots.  



```python
# EXPORT CSV FILE AND ORGANIZE DATA
# HEADERS ARE RENAMED TO HUMAN READABLE

URL  = "http://exoplanets.org/csv-files/exoplanets.1436203548.csv.gz"
data = pd.read_csv(URL, compression='gzip', delimiter=',',usecols=(0, 132, 145, 153, 168, 197))
data_table = data.columns = ['Semi_maj_ax','Plnt_mass','Star_mass','Plnt_num','Period','Star_rad']
data_clean = data.dropna()

# There are some rows that were removed. Possibly due to a systematic error.
# That is for rows with Planet Mass exactly equal to 1.
rowsToDrop = data.Plnt_mass == 1
data = data[rowsToDrop == False]


```

```python
#HERE WE ASSOCIATE THE RESPECTIVE COLUMNS IN "DATA", TO THE GIVEN VARIABLE NAMES BELOW
#VARIABLES HOLD A LIST OF DATA


semi_maj_ax = data['Semi_maj_ax']   # Semi major axis (astronomical units)
plnt_mass   = data['Plnt_mass']     # Planet Mass (Jupiter Mass)
star_mass   = data['Star_mass']     # Mass of Star (Solar masses)
plnt_num    = data['Plnt_num']      # Number of known planets in system
period      = data['Period']        # Orbital period (days)
star_rad    = data['Star_rad']      # Stelar Radius (Solar Radii) 
semi1       = semi_maj_ax           # ADDED THIS TO FIX A GLITCH
per1        = period                # SEE ABOVE
```

## Period vs Orbital Radius
Compare the number of days it takes an orbiting body to travel a complete rotation to the orbital radius. A zoomed-in graph is provided.

```python
# PLOT PERIOD VS ORBITAL RADIUS

plt.scatter(semi_maj_ax, period, label = "Experimental")
plt.xlabel('Orbital Radius (a.u.)')
plt.ylabel('Orbital Period (Days)')
plt.title('Transiting exoplanets - Orbital Radius vs Orbital Period')

# INSERT A FIT USING THEORETICAL EQ.
# T^2 = (4*pi^2)*(G*Mass_center)^-1 * a^3
# For Mass_center we will use an average of stellar masses from our data 

mass_center = np.nanmean(star_mass) # nanmean is a method(function) that ignores NaN (Not a Number) elements in a list
                                    # Units is Solar Masses
orb_rad = np.linspace(0,120,1000)
period  = ((4*(np.pi**2))*((G*mass_center)**-1)*(orb_rad**3))**0.5
plt.plot(orb_rad, period, c = 'red', label = "Theoretical")


plt.legend()
plt.show()



# THIS CAN BE IMPROVED TO CONVERT TO EARTH YEARS
# The fit is good for smaller masses. 

```

```python
# LET'S ZOOM IN!!!
# PLOT PERIOD VS ORBITAL RADIUS
x = semi1
y = per1
plt.scatter(semi1, per1, label = "Experimental")
plt.xlabel('Orbital Radius (a.u.)')
plt.ylabel('Orbital Period (Days)')
plt.title('Transiting exoplanets - Orbital Radius vs Orbital Period')

# INSERT A FIT USING THEORETICAL EQ.
# T^2 = (4*pi^2)*(G*Mass_center)^-1 * a^3
# For Mass_center we will use an average of stellar masses from our data 

mass_center = np.nanmean(star_mass) # nanmean is a method(function) that ignores NaN (Not a Number) elements in a list
                                    # Units is Solar Masses
orb_rad = np.linspace(0,120,1000)
period  = ((4*(np.pi**2))*((G*mass_center)**-1)*(orb_rad**3))**0.5
plt.plot(orb_rad, period, c = 'red', label = "Theoretical")


plt.legend()
plt.xlim(-0.2,14)
plt.ylim(-0.2,10e3)
plt.show()
```
